```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
from scipy import stats
```

```python
url = 'https://raw.githubusercontent.com/putriayudesita/kerja-praktik/main/RAW%20DATA%20SALARY%20OIL%20%26%20GAS%20COMPANY%20DUMMY.csv'
data = pd.read_csv(url)
data
```

```python
# Checking Missing Values
data.isnull().sum()
```

```python
data = data.drop(columns=['Unit Salary', 'Gross Salary', 'Converted Salary', 'Year', 'Start Date', 'End Date', 'Index', 'Location Type', 'Segment', 'Role Type', 'Education','Sumatra', 'Jakarta'])
data
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to all object-type columns
for column in data.select_dtypes(include=['object']).columns:
    data[column] = le.fit_transform(data[column])

# Display the DataFrame with encoded values
print(data)
```

```python
# Summary statistics for all columns
summary_stats = data.describe()
print(summary_stats)
```

```python
print(data.columns)
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# Add title and display the plot
plt.title('Correlation Heatmap')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Add a constant term for VIF calculation
X = add_constant(data.drop(columns=['Adjusted Salary']))

# Calculate VIF for each predictor
vif = pd.DataFrame()
vif['Variable'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
# Checking Outliers
def plot_boxplot(data, column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
plot_boxplot(data, 'Year of Experience')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the KNN Regressor model
knn = KNeighborsRegressor(n_neighbors=5, weights='distance', p=2)  # p=2 corresponds to Euclidean distance

# Train the model
knn.fit(X_train_scaled, y_train)

# Make predictions
y_pred = knn.predict(X_test_scaled)

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error on Test Data: {mae}")
```
